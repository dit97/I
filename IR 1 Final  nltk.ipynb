```python
import nltk 

from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
from nltk.stem import PorterStemmer 
```

```python
# Download the required NLTK datasets 
nltk.download('punkt') 
nltk.download('stopwords') 
nltk.download('punkt_tab')
```

```python
# Function for text preprocessing 
def preprocess_text(text): 
    # Tokenize the text 
    tokens = word_tokenize(text) 
    print("Tokenised words:", tokens)
    # Convert to lowercase 
    tokens = [token.lower() for token in tokens] 
 
    # Remove punctuation and non-alphabetic tokens 
    words = [word for word in tokens if word.isalpha()] 
 
    # Remove stopwords 
    stop_words = set(stopwords.words('english')) 
    words = [word for word in words if word not in stop_words] 
    print("Stop_words removed:", words)
 
    # Stemming using Porter Stemmer 
    ps = PorterStemmer() 
    stemmed_words = [ps.stem(word) for word in words] 
 
    return stemmed_words 
```

```python
# Example usage 
text = '''I am writing to express my keen interest in the Junior Data Analyst position at AlphaSense,
as advertised. As an undergrad in Data Science and hands-on experience in SQL, data
visualization, and analytics, I am excited about the opportunity to contribute to your
esteemed team and support AlphaSense's mission of leveraging data for growth.
AlphaSense's reputation as a market intelligence and search platform, coupled with its
commitment to innovation and growth, resonates deeply with me. The prospect of being
part of a team that utilizes cutting-edge technology to deliver actionable insights to leading
companies and financial institutions is truly inspiring.'''

```

```python
print("Original Text: ", text)
```

```python
processed_text = preprocess_text(text) 
```

```python
print("Processed Text:", processed_text) 
```
